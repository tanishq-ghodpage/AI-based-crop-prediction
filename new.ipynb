```python
import matplotlib.pyplot as plt  
import pandas as pd  

# Sample data (replace with your dataset)  
data = {  
    'Year': [2018, 2019, 2020, 2021, 2022],  
    'WestBengal_Rice': [3.8, 4.0, 4.2, 4.1, 4.3],  
    'Punjab_Wheat': [5.1, 5.3, 5.0, 5.4, 5.2]  
}  
df = pd.DataFrame(data)  

plt.figure(figsize=(10, 6))  
plt.plot(df['Year'], df['WestBengal_Rice'], marker='o', label='West Bengal (Rice)')  
plt.plot(df['Year'], df['Punjab_Wheat'], marker='s', label='Punjab (Wheat)')  
plt.title('Crop Yield Trends (2018-2022)', fontsize=14)  
plt.xlabel('Year', fontsize=12)  
plt.ylabel('Yield (tonnes/hectare)', fontsize=12)  
plt.grid(linestyle='--', alpha=0.7)  
plt.legend()  
```

```python
import seaborn as sns  

# Feature importance data  
features = ['Rainfall', 'Fertilizer', 'Pesticide', 'Area']  
importance = [0.35, 0.28, 0.20, 0.17]  

plt.figure(figsize=(8, 6))  
sns.barplot(x=importance, y=features, palette='viridis')  
plt.title('Feature Importance in Yield Prediction', fontsize=14)  
plt.xlabel('Relative Importance', fontsize=12)  
plt.ylabel('Features', fontsize=12)  
```

```python
crops = ['Rice', 'Wheat', 'Maize', 'Pulses']  
production = [45, 30, 15, 10]  

plt.figure(figsize=(8, 8))  
plt.pie(production, labels=crops, autopct='%1.1f%%',   
        colors=['#66bb6a', '#ffb74d', '#42a5f5', '#ec407a'],  
        startangle=90)  
plt.title('Crop Production Distribution', fontsize=14)  
```

```python
import numpy as np  

# Generate synthetic data  
np.random.seed(42)  
rainfall = np.random.normal(1200, 150, 100)  
yield_vals = 0.003 * rainfall + np.random.normal(0, 0.2, 100)  

plt.figure(figsize=(8, 6))  
sns.regplot(x=rainfall, y=yield_vals, line_kws={'color': 'red'})  
plt.title('Rainfall vs. Crop Yield Correlation', fontsize=14)  
plt.xlabel('Annual Rainfall (mm)', fontsize=12)  
plt.ylabel('Yield (tonnes/hectare)', fontsize=12)    
```

```python
from sklearn.metrics import mean_squared_error, r2_score  
import matplotlib.pyplot as plt  

# Sample predictions vs actual values  
y_true = [3.8, 4.0, 4.2, 4.1, 4.3]  # Actual yields  
y_pred = [3.7, 3.9, 4.3, 4.0, 4.2]  # Predicted yields  

mse = mean_squared_error(y_true, y_pred)  
r2 = r2_score(y_true, y_pred)  

print(f"MSE: {mse:.2f}")  
print(f"R²: {r2:.2f}")  

# Plot predictions vs actual  
plt.figure(figsize=(8, 6))  
plt.scatter(y_true, y_pred, c='#2e7d32', edgecolors='k')  
plt.plot([3.5, 4.5], [3.5, 4.5], 'r--')  # Perfect prediction line  
plt.title('Actual vs Predicted Yields', fontsize=14)  
plt.xlabel('Actual Yield (tonnes/ha)', fontsize=12)  
plt.ylabel('Predicted Yield (tonnes/ha)', fontsize=12)  
plt.grid(alpha=0.3)  
```

```python
import pandas as pd  

# Sample data  
years = [2018, 2019, 2020, 2021, 2022]  
yield_wb = [3.8, 3.9, 4.1, 4.0, 4.3]  

plt.figure(figsize=(10, 6))  
plt.plot(years, yield_wb, marker='o', color='#2e7d32', linestyle='-')  
plt.fill_between(years, yield_wb, alpha=0.1, color='#2e7d32')  
plt.title('Rice Yield Trends in West Bengal (2018–2022)', fontsize=14)  
plt.xlabel('Year', fontsize=12)  
plt.ylabel('Yield (tonnes/hectare)', fontsize=12)  
plt.xticks(years)  
plt.grid(linestyle='--', alpha=0.5)  
```

```python
import numpy as np  

fertilizer = np.linspace(0, 300, 50)  
yield_punjab = 5.0 - 0.005 * (fertilizer - 200)**2  # Quadratic relationship  

plt.figure(figsize=(8, 6))  
plt.plot(fertilizer, yield_punjab, color='#d84315')  
plt.axvline(x=200, color='#2e7d32', linestyle='--', label='Optimal Fertilizer (200 kg/ha)')  
plt.title('Wheat Yield vs Fertilizer Use in Punjab', fontsize=14)  
plt.xlabel('Fertilizer (kg/hectare)', fontsize=12)  
plt.ylabel('Yield (tonnes/hectare)', fontsize=12)  
plt.legend()  
plt.grid(alpha=0.3)  
```

```python
states = ['Punjab', 'Haryana', 'West Bengal', 'Uttar Pradesh', 'Tamil Nadu']  
yields = [5.4, 4.8, 4.3, 4.1, 3.9]  

plt.figure(figsize=(10, 6))  
bars = plt.bar(states, yields, color=['#2e7d32', '#a5d6a7', '#66bb6a', '#388e3c', '#1b5e20'])  
plt.bar_label(bars, fmt='%.1f', padding=3)  
plt.title('Top 5 States by Crop Yield (2022)', fontsize=14)  
plt.xlabel('State', fontsize=12)  
plt.ylabel('Yield (tonnes/hectare)', fontsize=12)  
plt.ylim(0, 6)  
plt.grid(axis='y', alpha=0.3) 
```
